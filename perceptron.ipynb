```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
#now give a input array 
X=np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
```

```python
y=np.array([0,0,0,1])
```

```python
w1=1.2
w2=0.6
weights=np.array([w1,w2])
threshold=1
bias=0
learning_rate=0.5
```

```python
for epoch in range(50):
    for i in range(len(X)):
        summation=np.dot(X[i],weights)+bias
        predicted_output=1 if summation>=1 else 0
        error=y[i]-predicted_output
        weights +=learning_rate * error *X[i]
        bias += learning_rate *error
print("trained weights :",weights,"bias :",bias)        
```

```python
#test perceptron 
for i in range(len(X)):
    z=np.dot(X[i],weights)+bias
    output=1 if z>=1 else 0
    print("inputs:",X[i] ,"output:",output)
    
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='cool')
x = np.linspace(-0.5, 1.5, 100)
y_boundary = -(weights[0] * x + bias) / weights[1]
plt.plot(x, y_boundary, '--k')
plt.title(f"and Gate Decision Boundary")
plt.xlabel("Input 1")
plt.ylabel("Input 2")
plt.grid(True)
plt.show()
```
